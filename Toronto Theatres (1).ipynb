```python
tort = pd.read_csv (r'toronto.csv')
print (tort)
```

```python
tort['LongAddress'] = tort[['Street Address', 'City', 'Prov']].agg(', '.join, axis=1)
print(tort)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=5)

# 2- - create location column
tort['Geo'] = tort['LongAddress'].apply(geocode)

# 3 - create longitude, laatitude and altitude from location column (returns tuple)
tort['LatLong'] = tort['Geo'].apply(lambda loc: tuple(loc.point) if loc else None)

# 4 - split point column into latitude, longitude and altitude columns
tort[['latitude', 'longitude', 'altitude']] = pd.DataFrame(tort['LatLong'].tolist(), index=tort.index)
```
